#### Write a function that will plot the ranking of a given roller coaster over time as a line.

```python
def rank_year(name, park):
  df = roll_wood[(roll_wood['Name'] == name) & (roll_wood['Park'] == park)]
  plt.figure(figsize = (10, 8))
  aa = plt.subplot()
  plt.plot(df['Year of Rank'], df['Rank'], color = 'red')
  plt.title(name)
  plt.xlabel('Year')
  plt.ylabel('Rank')
  aa.set_yticks([1, 2, 3, 4])
  aa.invert_yaxis()
  plt.legend([name], loc = 1)
  plt.show()

print(rank_year('El Toro', 'Six Flags Great Adventure'))
```

#### Write a function that will plot the ranking of two given roller coasters over time as lines.

```python
def rank_year2(name1, park1, name2, park2):
  df1 = roll_wood[(roll_wood['Name'] == name1) & (roll_wood['Park'] == park1)]
  df2 = roll_wood[(roll_wood['Name'] == name2) & (roll_wood['Park'] == park2)]
  plt.figure(figsize = (10, 8))
  ab = plt.subplot()
  plt.plot(df1['Year of Rank'], df1['Rank'], color = 'red')
  plt.plot(df2['Year of Rank'], df2['Rank'], color = 'green')
  plt.title(name1 + ' Vs ' + name2)
  plt.xlabel('Year')
  plt.ylabel('Rank')
  plt.legend([name1, name2], loc = 0)
  ab.set_yticks([1, 2, 3, 4])
  ab.invert_yaxis()
  plt.show()

print(rank_year2('El Toro', 'Six Flags Great Adventure', 'Boulder Dash', 'Lake Compounce'))
```

#### Write a function that will plot the ranking of the top n ranked roller coasters over time as lines.

```python
def top_rank(df3, n):
    top = df3[df3['Rank'] <= n]
    plt.figure(figsize = (10, 8))
    ac = plt.subplot()
        
    for coaster in set(top['Name']):
        coaster_rank = top[top['Name'] == coaster]
        ac.plot(coaster_rank['Year of Rank'], coaster_rank['Rank'], label = coaster)
        ac.set_yticks([i for i in range(1, 6)])
        
    plt.title("Top 10 Rankings")
    plt.xlabel("Year")
    plt.ylabel("Ranking")
    ac.invert_yaxis()
    plt.legend(loc = 4)
    plt.show()

print(top_rank(roll_wood, 5))
```

#### Create a function to plot histogram of column values.

```python
def hist_columns(df4, column):
    plt.figure(figsize = (10, 8))
    dfa = df4[column].dropna()
    plt.hist(dfa)
    legend = ([column.capitalize()])
    plt.legend(legend)
    plt.title("Roller Coasters by " + column.capitalize())
    plt.xlabel(column.capitalize())
    plt.ylabel('Roller Coasters Frequency')
    plt.show()
    
print(hist_columns(roll_coasters, 'speed'))

print(hist_columns(roll_coasters, 'length'))

print(hist_columns(roll_coasters, 'num_inversions'))
```

#### Write a function that creates a bar chart showing the number of inversions for each roller coaster at an amusement park.

```python
def hist_height(df5, column):
  if column == 'height':
    dfb = df5[df5[column] <= 140]
    dfc = dfb.dropna()
  else:
    dfc = dfb.dropna()

  df6 = dfc[column]
  plt.figure(figsize = (10, 8))
  plt.hist(df6)
  legend = ([column.capitalize()])
  plt.legend(legend)
  plt.title("Roller Coasters by " + column.capitalize())
  plt.xlabel(column.capitalize())
  plt.ylabel('Roller Coasters Frequency')
  plt.show()

print(hist_height(roll_coasters, 'height'))
```

#### Create a function to plot inversions by coaster at park

```python
def bar_inversions(df7, park):
  park_coasters = df7[(df7['park'] == park) & (df7['num_inversions'] > 0)]
  park_coasters = park_coasters.sort_values('num_inversions', ascending = False)
  coaster_names = park_coasters['name']
  number_inversions = park_coasters['num_inversions']
  plt.figure(figsize = (10, 8))
  ad = plt.subplot()
  plt.bar(range(len(coaster_names)), number_inversions)
  ad.set_xticks(range(len(coaster_names)))
  ad.set_xticklabels(coaster_names)
  plt.title(f"Number of Roller Coaster Inversions in {park} amusement park")
  plt.xlabel("Roller Coaster")
  plt.ylabel("Number of Inversions")
  plt.show()

print(bar_inversions(roll_coasters, 'Parc Asterix'))
```

#### Create a function to plot a pie chart of status.operating

```python
def pie_chart_operating(df8):
  operating_coasters = roll_coasters[roll_coasters['status'] == 'status.operating']
  closed_coasters = roll_coasters[roll_coasters['status'] == 'status.closed.definitely']
  status_counts = [len(operating_coasters), len(closed_coasters)]
  ops_labels = ['Operational', 'Closed']

  plt.clf()
  plt.pie(status_counts, autopct = '%0.1f%%', labels = ops_labels)
  plt.title('Operational Status Proportion')
  plt.axis('equal')
  plt.show()
  
print(pie_chart_operating(roll_coasters))
```

#### Create a function to plot scatter of any two columns

```python
plt.clf()
def scatter_plot_two(df9, col1, col2):
  df_col1 = df9[col1]
  df_col2 = df9[col2]
  plt.scatter(df_col1, df_col2, marker = 'o')
  plt.xlabel(col1)
  plt.ylabel(col2)
  plt.title('Scatter Plot of ' + col2 + ' against ' + col1 + '.')
  plt.show()

print(scatter_plot_two(roll_coasters, 'length', 'speed'))
```

#### Create a function to plot scatter of speed vs height

```python
plt.clf()
def scatter_plot_height_speed(df10, col3, col4):
  if col3 == 'height':
    df10 = df10[df10[col3] <= 140]
    df11 = df10.dropna()
  elif col4 == 'height':
    df10 = df10[df10[col4] <= 140]
    df11 = df10.dropna()
  else:
    df11 = df10.dropna()

  df_col3 = df11[col3]
  df_col4 = df11[col4]

  plt.scatter(df_col3, df_col4, marker = 'o')
  plt.xlabel(col3)
  plt.ylabel(col4)
  plt.title('Scatter Plot of ' + col3 + ' against ' + col4 + '.')
  plt.show()

print(scatter_plot_height_speed(roll_coasters, 'height', 'speed'))
```
